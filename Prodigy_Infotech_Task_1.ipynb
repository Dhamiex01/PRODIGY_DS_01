```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("worldpopulationdata.csv")
data.head()
```

```python
print(data["Series Name"].unique())
print(data["Series Code"].unique())
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.describe()
```

###DATA VISUALIZATION WITH RESPECT TO THE TOTAL POPULATION

```python
total_pop_data = data[data["Series Code"] == 'SP.POP.TOTL']
total_pop_data
```

```python
###Extraction of the top 10 countries wrt to the total population
total_pop_data_sorted = total_pop_data.sort_values(by = "2022", ascending = False)
total_pop_data_top_15 = total_pop_data_sorted.head(15)
total_pop_data_top_15
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.barplot(data = total_pop_data_top_15, y = "2022", x = "Country Name", palette = "viridis")
plt.xlabel("Country Names")
plt.ylabel("Income")
plt.xticks(rotation = 90)
plt.title("Graph Showing the top 10 Countries with the highest count w.r.t Total Population")
plt.show()
```

```python
###Extraction of the buttom 10 countries wrt to the total population
total_pop_data_sorted1 = total_pop_data.sort_values(by = "2022", ascending = True)
total_pop_data_buttom_15 = total_pop_data_sorted1.head(15)
total_pop_data_buttom_15
```

```python
sns.barplot(data = total_pop_data_buttom_15, y = "2022", x = "Country Name", palette = "viridis")
plt.xlabel("Country Names")
plt.ylabel("Income")
plt.xticks(rotation = 90)
plt.title("Graph Showing the top 10 Countries with the lowest count w.r.t Total Population")
plt.show()
```
